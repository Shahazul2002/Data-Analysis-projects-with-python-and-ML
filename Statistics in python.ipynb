```python
ages = [21, 25, 19, 19, 25, 22, 24, 24, 27]
```

```python
mean_age = sum(ages) / len(ages)
mean_age
```

```python
sorted_ages = sorted(ages)
n = len(sorted_ages)
median_age = (sorted_ages[n//2] if n%2 != 0 else (sorted_ages[n//2 - 1] + sorted_ages[n/2]) / 2)
print(sorted_ages)
print(f"The median age is {median_age}")
```

```python
from collections import Counter
mode_age = Counter(ages)
max_frequency = max(mode_age.values())
modes = [k for k, v in mode_age.items() if v == max_frequency]
modes
```

Measure of skewness:


```python
#Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew
import seaborn as sns
```

```python
#Generating Data
data1 = np.random.normal(loc=0, scale=1, size=1000) #Normally distributed data
skewness = skew(data1)
print(f"Skewness: {skewness}")
```

Interpretation:
#Near zero skewness:- The value of skewness is very close to zero, which implies that the characteristics of a symmetric distribution similar to a normal distribution.
#Slight negative value:- The slight negative value suggests a very minor skew to the left. However, this is so close to zero that for most practical purposes the distribution can be considered as symmetric.

```python
sns.histplot(data=data1, kde=True)
```

```python
data2 = pd.DataFrame({'values': np.random.lognormal(mean=0, sigma=1, size=1000)}) #Lognormal distribution
skewness = data2['values'].skew()
skewness
```

Interpretation:
    The value of skewness is 5.5 aproximately which implies that the distribution is positively skewed.

```python
sns.histplot(data=data2, kde=True)
```

```python
 data = np.random.lognormal(mean=0, sigma=1, size=1000)
```

```python
mean = np.mean(data)
median = np.median(data)
mode = max(set(data), key=list(data).count)
```

```python
plt.figure(figsize=(10, 4))
plt.hist(data, bins=30, alpha=0.6, color='blue', label='Data')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean:.2f}")
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f"Median: {median:.2f}")
plt.axvline(mode, color='blue', linestyle='dashed', linewidth=2, label=f"Mode: {mode:.2f}")

plt.title("Right Skewed Distribution")
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
skewness = skew(data)
print(f"Skewness: {skewness}")
```

Comment: The value of skewness is 3.29 which implies that the distribution is positively skewed.

```python
data = np.random.normal(loc=0, scale=1, size=1000)

mean = np.mean(data)
median = np.median(data)
mode = max(set(data), key=list(data).count)

plt.figure(figsize=(8, 4))
plt.hist(data, bins=30, alpha=0.6, color='blue', label='Data')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean:.2f}")
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f"Median: {median:.2f}")
plt.axvline(mode, color='blue', linestyle='dashed', linewidth=2, label=f"Mode: {mode:.2f}")

plt.title("Symmetric Distribution")
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
