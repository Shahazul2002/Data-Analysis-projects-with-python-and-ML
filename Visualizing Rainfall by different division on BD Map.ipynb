```python
# 1. Load Bangladesh shapefile (divisions)
# Replace with your shapefile or GeoJSON file
gdf = gpd.read_file("bd.json")

# 2. Load dataset (division-wise statistics, e.g., rainfall)
data = {
    "Division": ["Dhaka", "Chittagong", "Khulna", "Rajshahi", "Barisal", "Sylhet", "Rangpur", "Mymensingh"],
    "Rainfall": [2100, 3500, 1800, 1600, 2500, 4000, 1700, 3000]
}
df = pd.DataFrame(data)

# 3. Merge shapefile with dataset
merged = gdf.merge(df, left_on="name", right_on="Division")

# 4. Plot map
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
merged.plot(column="Rainfall",
            cmap="Blues",   # color scheme
            linewidth=0.8,
            edgecolor="black",
            legend=True,
            ax=ax)

# Add title
plt.title("Bangladesh Division-wise Rainfall by Shahazul", fontsize=15)
plt.axis("off")
plt.show()
```

```python
# Plot
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
merged.plot(column="Rainfall", cmap="Blues", edgecolor="black", legend=True, ax=ax)

# Add labels at centroid of each division
for idx, row in merged.iterrows():
    plt.annotate(text=row["Division"], 
                 xy=(row.geometry.centroid.x, row.geometry.centroid.y),
                 horizontalalignment="center", fontsize=10, color="black")

plt.title("Bangladesh Division-wise Rainfall map by Shahazul", fontsize=16)
plt.axis("off")
##Save the map
#plt.savefig("bangladesh_rainfall_map.png", dpi=300, bbox_inches="tight")  # PNG
# plt.savefig("bangladesh_rainfall_map.jpg", dpi=300, bbox_inches="tight")  # JPG
# plt.savefig("bangladesh_rainfall_map.pdf", dpi=300, bbox_inches="tight")  # PDF

plt.show()
```
